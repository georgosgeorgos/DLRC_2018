```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
sns.set(style="white", palette="muted", color_codes=True)
```

```python
lidar_data = np.load("lidar_data.npy")
```

```python
lidar_data = lidar_data.astype(float)
```

```python
lidar_data[lidar_data > 1900] = -1
```

```python
(lidar_data == -1).mean()
```

```python
plt.figure(figsize=(20,10))
plt.plot(lidar_data[:, 0])
plt.grid()
```

```python
plt.figure(figsize=(20,10))
plt.hist(lidar_data[:, 0])
plt.grid()
```

```python
sample_run0 = lidar_data[100:200,:]
sample_run1 = lidar_data[1100:1200,:]
```

```python
# Set up the matplotlib figure
f, axes = plt.subplots(9, 2, figsize=(40, 40), sharex=True)
sns.despine(left=True)

for j in range(9):
    try:
        # Plot a historgram and kernel density estimate
        sns.distplot(sample_run0[:,j], color="g", ax=axes[j, 0])
    except np.linalg.LinAlgError:
        continue
    try:
        sns.distplot(sample_run1[:,j], color="b", ax=axes[j, 1])
    except np.linalg.LinAlgError:
        continue

    plt.setp(axes, yticks=[])
    plt.tight_layout()
```

```python
# Set up the matplotlib figure
f, axes = plt.subplots(9, 1, figsize=(40, 40), sharex=True)
sns.despine(left=True)

for j in range(9):
    sns.distplot(lidar_data[:,j], color="b", ax=axes[j])

plt.setp(axes, yticks=[])
plt.tight_layout()
```
