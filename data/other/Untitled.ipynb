```python
import pickle as pkl
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open("./cluster.pkl", "rb") as f:
    data = pkl.load(f)
```

```python
lidar = [data[i]["lidar"]["measurements"] for i in range(3)]
```

```python
lidar = np.array(lidar)
```

```python
lidar[lidar > 2000] = 2000
```

```python
plt.figure(figsize=(20,10))
plt.plot(lidar[0][:,3])
plt.grid()
plt.show()
```
