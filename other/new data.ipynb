```python
import numpy as np
from PIL import Image
import os
import json
import matplotlib.pyplot as plt
import pickle as pkl
```

```python
with open("./data_0.pkl", "rb") as f:
    data = pkl.load(f)[0]
```

```python
state = data["state"]
lidar = data["lidar"]
trajectory = data["trajectory"]
```

```python
lidar = lidar["measurements"]
```

```python
lidar = np.array(lidar).astype(float)
```

```python
lidar[lidar > 2000] = 2000
```

```python
lidar = lidar / 1000
```

```python
f, ax = plt.subplots((lidar.shape[1]), figsize=(20,40))

for i in range(lidar.shape[1]):
    ax[i].plot(lidar[:1000, i])
    ax[i].set_xlabel("time")
    ax[i].set_xlabel("lidar "+str(i))
plt.show()
```
